```python
animeler=df.values
print(animeler[3])
```

```python
word2idx = {}
i = 0

for anime in animeler:
    for sozcuk in str(anime).split():
        if sozcuk not in word2idx:
            word2idx[sozcuk] = i
            i += 1
```

```python
word2idx
```

```python
anime_sayisi = len(animeler)
sozluk_buyuklugu = len(word2idx)

print(anime_sayisi, sozluk_buyuklugu)
```

```python
tf = np.zeros((anime_sayisi, sozluk_buyuklugu))

for i, anime in enumerate(animeler):
    sozcukler = str(anime).split()
    for sozcuk in sozcukler:
        tf[i, word2idx[sozcuk]] += 1

    tf[i] = tf[i] / len(sozcukler)

print(tf[0])
```

```python
toy_tf = np.array([
    [0.5, 0.0, 0.0, 0.5],
    [0.0, 0.0, 0.5, 0.5],
    [0.0, 0.1, 0.0, 0.9],
    [0.3, 0.0, 0.1, 0.6],
    [0.7, 0.0, 0.0, 0.3],
])

toy_doc_freq = np.sum(toy_tf > 0, axis=0)
toy_doc_freq
```

```python
toy_idf = 5 / toy_doc_freq
print(toy_idf)
toy_idf = np.log(toy_idf)
print(toy_idf)
```

```python
doc_freq = np.sum(np.zeros((anime_sayisi, sozluk_buyuklugu))> 0, axis=0)
print(doc_freq)

idf = np.log(anime_sayisi / doc_freq)
print(idf)
```

```python
tf_idf = tf * idf
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
def film2str(row_index):
    anime = df.iloc[row_index]
    adi = anime["name"]
    turu = "".join(anime["type"])
    ozet = re.sub(r"\n+", "\n", anime["episodes"]).strip()
    if len(ozet) > 200:
        ozet = ozet[:200] + "..."

    return f"Anime adı: {adi} \nTürü: {turu} \nBölüm sayısı: {ozet}"


print(film2str(3))
```

```python
def benzerlerini_bul(anime_adi, benzer_anime_sayisi):
    anime_indexi = df[df["name"] == anime_adi].index[0]
    print("Aranan anime:")
    print(film2str(anime_indexi))
    print()
    print("* " * 30)
    print()

    benzerlikler = []
    sorgu_film = tf_idf[anime_indexi]
    for vec in tf_idf:
        benzerlik = cos_sim(sorgu_film, vec)
        benzerlikler.append(benzerlik)

    en_benzer_anime_indexleri = (- np.array(benzerlikler)
                                ).argsort()[:benzer_anime_sayisi + 1]
    print("Benzer Filmler:\n")
    for i, anime_idx in enumerate(en_benzer_anime_indexleri[1:]):
        print(str(i+1) + ") " + film2str(anime_idx))
        print()
        print("- " * 20)
        print()
```
